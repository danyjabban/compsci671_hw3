## 6 Ridge Regression and its friends {-}

### 6.1 {-}

```python
import numpy as np
data = np.random.normal(3, 2.5, size=(100, 5000000))
```

```python
from sklearn.linear_model import Ridge

rr = Ridge(alpha=1)
rr.fit(data[:, 1:], data[:, 0])
```

```python
from sklearn.kernel_ridge import KernelRidge

krr = KernelRidge(alpha=1)
krr.fit(data[:, 1:], data[:, 0])
```

### 6.2 {-}

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

data2 = np.random.normal(0, .70, size=(1000, 500))
X = data2[:, 1:]
y = data2[:, 0]

rr = Ridge(alpha=.5)
rr.fit(X, y)
lr = Lasso(alpha=.5)
lr.fit(X, y)
```

### 6.3 {-}

```python
import matplotlib.pyplot as plt
reg_strength = np.arange(0.01, 0.1, .01)

# lasso
coef_mat = np.array([])
for alpha in reg_strength:
    lr = Lasso(alpha=alpha)
    lr.fit(X, y)
    coef = lr.coef_
    coef_mat = np.append(coef_mat, coef)
   
coef_mat = coef_mat.reshape(reg_strength.shape[0], X.shape[1])
inv_reg_strength = np.array([])
for i in reg_strength:
    inv_reg_strength = np.append(inv_reg_strength, 1/i)
for i in range(coef_mat.shape[1]):
    plt.plot(inv_reg_strength, coef_mat[:, i])

plt.title("Lasso Regression coefficients vs Log Inverse Regularization Strength")
plt.show()
```

```python
reg_strength = np.arange(0.01, 0.1, .01)
# ridge
coef_mat = np.array([])
for alpha in reg_strength:
    rr = Ridge(alpha=alpha)
    rr.fit(X, y)
    coef = rr.coef_
    coef_mat = np.append(coef_mat, coef)
    
coef_mat = coef_mat.reshape(reg_strength.shape[0], X.shape[1])
inv_reg_strength = np.array([])
for i in reg_strength:
    inv_reg_strength = np.append(inv_reg_strength, 1/i)
for i in range(coef_mat.shape[1]):
    plt.plot(inv_reg_strength, coef_mat[:, i])

plt.title("Ridge Regression coefficients vs Log Inverse Regularization Strength")
plt.show()

```
